```python
df = orders_df.copy()
df = df[df["order_status"].eq("delivered")].copy()
df = df.dropna(subset=["order_delivered_customer_date","order_estimated_delivery_date"])

df["month"] = df["order_purchase_timestamp"].dt.to_period("M")
df["is_late"] = df["order_delivered_customer_date"] > df["order_estimated_delivery_date"]

ontime_monthly = (
    df.groupby("month")
    .agg(n_delivered=("order_id","size"),
         n_late=("is_late","sum"))
    .reset_index()
)

ontime_monthly["late_rate"] = ontime_monthly["n_late"] / ontime_monthly["n_delivered"] * 100
ontime_monthly["month"] = ontime_monthly["month"].astype(str)
ontime_monthly.tail(12)
```

### Pertanyaan 1 : Bagaimana performa penjualan dan revenue perusahaan dalam beberapa bulan terakhir?

```python
# Kolom yang dipakai:
# - order_id
# - order_status
# - order_purchase_timestamp
# - payment_value

data = all_data.copy()

data["order_purchase_timestamp"] = pd.to_datetime(data["order_purchase_timestamp"], errors="coerce")
data["payment_value"] = pd.to_numeric(data["payment_value"], errors="coerce")

data = data[data["order_status"].eq("delivered")].copy()

data["month"] = data["order_purchase_timestamp"].dt.to_period("M")

order_level = (
    data.groupby(["month", "order_id"], as_index=False)
        .agg(order_revenue=("payment_value", "sum"))
)

monthly_perf = (
    order_level.groupby("month", as_index=False)
        .agg(
            n_delivered_orders=("order_id", "nunique"),
            revenue=("order_revenue", "sum"),
            aov=("order_revenue", "mean"),
        )
        .sort_values("month")
)

last_n = 12
monthly_last = monthly_perf.tail(last_n).copy()
monthly_last["month"] = monthly_last["month"].astype(str)

monthly_last
```

```python
plt.figure(figsize=(10,4))
plt.plot(monthly_last["month"], monthly_last["n_delivered_orders"], marker="o")
plt.title("Tren Penjualan (Jumlah Delivered Orders) - 12 Bulan Terakhir")
plt.xlabel("Bulan")
plt.ylabel("Jumlah Delivered Orders")
plt.xticks(rotation=45, ha="right")
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

#### Tren jumlah delivered orders selama 12 bulan terakhir terlihat fluktuatif. Jumlah order naik dari sekitar 4.1 ribu pada 2017-09 ke puncak sekitar 7.3 ribu pada 2017-11, lalu turun tajam pada 2017-12 ke sekitar 5.5 ribu. Setelah itu, volume kembali tinggi pada 2018-01 sekitar 7.1 ribu dan relatif stabil di kisaran 6.6 ribu sampai 7.0 ribu sepanjang 2018-02 hingga 2018-05. Pada 2018-06 terjadi penurunan ke sekitar 6.1 ribu, lalu mulai naik lagi secara bertahap pada 2018-07 dan 2018-08 hingga sekitar 6.4 ribu. Intinya, performa penjualan cenderung kuat pada akhir 2017 hingga awal 2018, sempat melemah di pertengahan 2018, lalu menunjukkan pemulihan di dua bulan terakhir pada grafik.

```python
plt.figure(figsize=(10,4))
plt.plot(monthly_last["month"], monthly_last["revenue"], marker="o", color="tab:green")
plt.title("Tren Revenue (Delivered Orders) - 12 Bulan Terakhir")
plt.xlabel("Bulan")
plt.ylabel("Revenue (Total Payment Value)")
plt.xticks(rotation=45, ha="right")
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

#### Tren revenue 12 bulan terakhir terlihat mengikuti pola yang mirip dengan tren jumlah delivered orders. Revenue meningkat dari 2017-09 dan mencapai puncak di sekitar 2017-11, lalu turun cukup tajam pada 2017-12. Setelah itu revenue kembali tinggi pada 2018-01, kemudian relatif stabil pada 2018-02 sampai 2018-05. Pada 2018-06 terjadi penurunan, lalu revenue mulai naik lagi pada 2018-07 dan 2018-08. Artinya, performa revenue perusahaan sempat sangat kuat di akhir 2017 dan awal 2018, melemah di pertengahan 2018, dan menunjukkan pemulihan pada dua bulan terakhir. Jika dibandingkan dengan grafik jumlah order, perbedaan naik-turun revenue vs order bisa dipakai untuk membaca perubahan nilai transaksi rata-rata per order (AOV).

```python
plt.figure(figsize=(10,4))
plt.plot(monthly_last["month"], monthly_last["aov"], marker="o", color="tab:orange")
plt.title("Tren AOV (Average Order Value) - 12 Bulan Terakhir")
plt.xlabel("Bulan")
plt.ylabel("AOV")
plt.xticks(rotation=45, ha="right")
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

#### Grafik tersebut menunjukkan tren nilai transaksi rata-rata per order (average order value atau AOV) selama 12 bulan terakhir. AOV terlihat berubah-ubah dari bulan ke bulan, yang artinya kenaikan atau penurunan revenue tidak selalu sejalan dengan jumlah order. Saat AOV naik, perusahaan menghasilkan revenue lebih besar per order, biasanya karena lebih banyak pembelian produk berharga tinggi atau keranjang belanja (basket) yang lebih besar. Saat AOV turun, revenue per order mengecil meskipun jumlah order bisa saja tetap tinggi, yang mengindikasikan transaksi didominasi produk dengan harga lebih rendah atau pembelian dengan nilai kecil lebih banyak.

# **Analisis Lanjutan (RFM Analysis)**

## Untuk menjawab tiga pertanyaan analisis terakhir, kita bisa menggunakan teknik analisis lanjutan yang bernama RFM analysis. Sederhananya, RFM analysis merupakan salah satu metode yang umum digunakan untuk melakukan segmentasi pelanggan (mengelompokkan pelanggan ke dalam beberapa kategori) berdasarkan tiga parameter, yaitu recency, frequency, dan monetary.

- ## Recency: parameter yang digunakan untuk melihat kapan terakhir seorang pelanggan melakukan pembelian.
- ## Frequency: parameter ini digunakan untuk mengidentifikasi seberapa sering seorang pelanggan melakukan pembelian.
- ## Monetary: parameter terakhir ini digunakan untuk mengidentifikasi seberapa besar revenue yang berasal dari uang yang dikeluarkan pelanggan tersebut.

```python
data = all_data.copy()
data["payment_value"] = pd.to_numeric(data["payment_value"], errors="coerce")
data = data[data["order_status"].eq("delivered")].copy()

order_level = (
    data.groupby(["customer_unique_id", "order_id"], as_index=False)
        .agg(
            order_date=("order_purchase_timestamp", "max"),
            order_value=("payment_value", "sum"),
        )
        .dropna(subset=["customer_unique_id", "order_id", "order_date", "order_value"])
)

analysis_date = order_level["order_date"].max().normalize() + pd.Timedelta(days=1)

recency = (
    order_level.groupby("customer_unique_id", as_index=False)
        .agg(last_purchase=("order_date", "max"))
)
recency["recency_days"] = (analysis_date - recency["last_purchase"].dt.normalize()).dt.days

TOP_N = 15
top_recency = (
    recency.nsmallest(TOP_N, "recency_days")
           .sort_values("recency_days", ascending=True)
)

X_MONTHS = 6
start_date = analysis_date - pd.DateOffset(months=X_MONTHS)

last6 = order_level[order_level["order_date"] >= start_date].copy()

rfm_6m = (
    last6.groupby("customer_unique_id", as_index=False)
         .agg(
             frequency=("order_id", "nunique"),
             monetary=("order_value", "sum"),
         )
)

top_frequency = (
    rfm_6m.nlargest(TOP_N, "frequency")
          .sort_values("frequency", ascending=True)
)

top_monetary = (
    rfm_6m.nlargest(TOP_N, "monetary")
          .sort_values("monetary", ascending=True)
)


# VISUALISASI (3 barplot)

fig, axes = plt.subplots(1, 3, figsize=(18, 5))

# Recency
axes[0].barh(top_recency["customer_unique_id"], top_recency["recency_days"], color="tab:blue")
axes[0].set_title(f"Top {TOP_N} Paling Baru Transaksi (Recency)")
axes[0].set_xlabel("Recency (hari sejak transaksi terakhir)")
axes[0].set_ylabel("customer_unique_id")
axes[0].invert_yaxis()

# Frequency
axes[1].barh(top_frequency["customer_unique_id"], top_frequency["frequency"], color="tab:green")
axes[1].set_title(f"Top {TOP_N} Paling Sering Beli (Frequency) - {X_MONTHS} Bulan")
axes[1].set_xlabel("Jumlah order (6 bulan terakhir)")
axes[1].set_ylabel("customer_unique_id")

# Monetary
axes[2].barh(top_monetary["customer_unique_id"], top_monetary["monetary"], color="tab:orange")
axes[2].set_title(f"Top {TOP_N} Pengeluaran Terbesar (Monetary) - {X_MONTHS} Bulan")
axes[2].set_xlabel("Total pengeluaran (payment_value) 6 bulan terakhir")
axes[2].set_ylabel("customer_unique_id")

plt.tight_layout()
plt.show()
```

#### Pada grafik RFM di atas dapat diketahui bahwa :
- #### Pada grafik Recency (Top 15 pelanggan yang paling baru melakukan transaksi), terdapat 11 pelanggan yang paling baru (1 hari) melakukan transaksi pembelian, 4 pelanggan lainnya baru melakukan transaksi dalam 2 hari terakhir.
- #### Pada grafik Frequency (Top 15 pelanggan yang paling sering lakukan pembelian selama 6 bulan terakhir), terdapat sebanyak 1 pelanggan yang melakukan transaksi lebih dari 6 kali, kemudian disusul oleh 5 pelanggan yang melakukan pembelian berulang sebanyak 4 kali, dan sisanya sebanyak 9 pelanggan melakukan pembelian sebanyak sekitar 3 kali.
- #### Pada grafik Monetary ( Top 15 pelanggan dengan pengeluaran terbesar selama 6 bulan terakhir), dapat diketahui bahwa pelanggan dengan total pengeluaran tertinggi yaitu sekitar 30000 dollar, diikuti oleh pelanggan-pelanggan lainnya yang total pengeluarannya sudah akan mencapai 30000 dollar. Pelanggan urutan terakhir melakukan pembelian dengan total sekitar 9000 dollar.
